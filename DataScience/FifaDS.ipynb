```python
df2 = pd.DataFrame(data_frame, columns = ['Name', 'Wage', 'Value'])
```

```python
df2
```

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

```python
wage = df2["Wage"].replace('[\€]', '', regex=True).apply(value_to_float)
value = df2["Value"].replace('[\€]', '', regex=True).apply(value_to_float)
```

```python
df2["Wage"] = wage
df2["Value"] = value

df2['Difference'] = df2["Value"]-df2["Wage"]

df2.head()
```

```python
df2.sort_values('Difference', ascending=False)
```
